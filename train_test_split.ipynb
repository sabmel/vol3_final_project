## Example of the Train-Test Split Code

```python
import kagglehub
from football.data_loader import DataLoader

path = kagglehub.dataset_download("maxhorowitz/nflplaybyplay2009to2016")
```

```python
dl = DataLoader(path)
```

```python
print(dl)
```

```python
print(dl[0])
```

```python
dl[0][0].clean()
dl[0][0]
```

```python
off_train, off_test = dl[0][0].train_test_split(offense=True)
def_train, def_test = dl[0][0].train_test_split(offense=False)
all_train, all_test = dl[0][0].train_test_split()

# Ensure that offensive and defensive sets are exhaustive
print(len(all_train), len(all_test))
print(len(off_train), len(off_test))
print(len(def_train), len(def_test))
```
